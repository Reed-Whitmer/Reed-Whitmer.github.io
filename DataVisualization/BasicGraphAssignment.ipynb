```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
###Import dataset
df_opiod = pd.read_csv('C:/Users/reedw\\CTBA\\CTBA\\M2\\OpidsVA.csv', header = 0)
###print(df_opiod)
```

Question 1:

Do opioid overdoes tend to be associated with less affluent areasâ€”that is, areas where
families have lower incomes?


```python
###Create the scatter plot
fig, ax = plt.subplots()

ax.scatter(df_opiod['MedianHouseholdIncome'], df_opiod['FPOO-Rate'], alpha = 0.5, c = 'r')
fig.suptitle('Opiod Overdoses vs Household Income')   
ax.xaxis.set_label_text('Median Household Income')  
ax.yaxis.set_label_text('Opiod Overdose per 100,000 Residents')
fig.set_size_inches(10,5)
plt.show()
```

I chose a scatter plot because the points will allow the user to make a comparison between the whole set of data on opiod overdose and their income.

Question 2:


Some people who start with opioid addictions are reported to transition to heroin use.
What is the relationship in Virginia counties between opioid overdoses and heroin
overdoses?

```python
###Create the double bar graph
labels = df_opiod['CountyName']
x = np.arange(len(labels))
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, df_opiod['FFHO-Rate'] , width, label='Heroine')
rects2 = ax.bar(x + width/2, df_opiod['FPOO-Rate'], width, label='Opiod')

fig.suptitle('Opiod Overdoses vs Heroine Overdoses by county')
ax.xaxis.set_label_text('Counties')  
ax.yaxis.set_label_text('Overdoses per 100,000 residents')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.autoscale(tight=True)
plt.xticks(rotation = 90)
fig.set_size_inches(40,5)

plt.show()
```

I chose a doublebar graph because the user will be able to check each county at the same time and look at the direct comparison between heroin and opiods overdose.

```python
### import second dataset
df_presidents = pd.read_csv('C:/Users/reedw\\CTBA\\CTBA\\M2\\presidents.csv', header = 0)
###print(df_presidents)
```

Question 3:

Which states are associated with the greatest number of United States presidents in
terms of the presidentsâ€™ birthplaces?


```python
### Count each state value 
states = df_presidents['State']
states = dict(states.value_counts())

###Create the bargraph
fig, ax = plt.subplots()


plt.bar(range(len(states)), list(states.values()), align = 'center', color = 'b')
plt.xticks(range(len(states)), list(states.keys()))
plt.xticks(rotation = 90)

fig.suptitle('Frequency of Presidents by State')
ax.xaxis.set_label_text('States')  
ax.yaxis.set_label_text('Frequency')


plt.show()
```

I chose this type of graph because it would allow the user to see, at the left of the graph, which state has the most presidents.  

```python
###Import third dataset
df_vehicle = pd.read_csv('C:/Users/reedw\\CTBA\\CTBA\\M2\\TOTALNSA.csv', header = 0)
##print(df_vehicle)
```

Question 4:

How have vehicle sales in the United States varied over time?

```python
#Remove the month and day, sum each year 
totals = df_vehicle.groupby(df_vehicle.DATE.astype(str).str[:4]).sum().reset_index()
###print(totals)

### Create the plot
fig, ax = plt.subplots()

plt.plot(totals['DATE'], totals['TOTALNSA'], color = 'b', marker = 'o')

plt.title('Total Sales per Year', fontsize=20)
plt.xlabel('Year', fontsize=20)
plt.ylabel('Total Sales', fontsize=20)
plt.grid(True)
plt.ylim(0, 20000)
fig.set_size_inches(30,10)
plt.show()


```

This type of graph will allow the user to view, at a glance, the trend of vehicle sales throughout the years.
